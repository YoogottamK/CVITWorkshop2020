```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("images/objects.png", cv.IMREAD_GRAYSCALE)

plt.imshow(img, cmap="gray")
```

```python
kernel = np.ones((3, 3))
```

```python
eroded = cv.erode(img, kernel)

plt.imshow(eroded, cmap="gray")
```

```python
hom = cv.morphologyEx(img, cv.MORPH_HITMISS, kernel)

plt.imshow(hom, cmap="gray")
```

```python
(hom == eroded).all()
```
